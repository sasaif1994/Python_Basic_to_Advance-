# <center> One-Way ANOVA

#### Question:

A firm wishes to compare 4 programmers for trainning workers to perform a certain task.20 new employees are randomly assigned to theses programmers with 5 in each.at the end of the tarinning , a test is conducted to see how quickly the trainees perform the task.the number of times the task is performed per hour is recorded for each trainee
    
Perform ANOVA to check the effectiveness of the programme.(Assume the level of significance=0.05)

#### Solutions:

Let µ1 be the number of times the task is performed per hour by trainee in the first programme

Similarly, µ2,µ3,µ4 be the number of times the task is performed per hour by trainee in the second , third and foruth programme respectively.

The test hypothesis is :

      Ho:The average number of times the task is performed are equal, i.e, µ1=µ2=µ3=µ4

##### Against
     Ha:The average number of times the task is performed is not equal, i.e,µ1≠µ2≠µ3≠µ4


```python
from IPython.display import Image
Image(filename='D:/Imarticus/Probablity & Statics/analysis of variance/one_way anova sol.PNG',width=400, height=200)
```

```python
# Import the required library
from scipy.stats import f_oneway

# Write the data
# data from programme 1
programme_1 =[9,12,14,11,13]

#data from programme 2
programme_2=[10,6,9,9,10]

#data from programme 3
programme_3=[12,14,11,13,11]

#data from programme 4
programme_4=[9,8,11,7,8]

#perform the test
f_oneway(programme_1,programme_2,programme_3,programme_4)
```

# <center> Chi-Square Test
 * Test are non-parametric
 * Do not follow any specific distribution & no assumption is made about the population.

### Test for independence of attrebutes

#### Question:

The data gives the distribution of students according to the type of music they prefer and their IQ level

```python
from IPython.display import Image
Image(filename='D:/Imarticus/Probablity & Statics/analysis of variance/Chi_square test_question.PNG',width=400, height=200)
```

Test whether the type of preferred music and IQ level are independent.

#### Solutions:
Ho:The attributes IQ level and music are independent of each other.
 
 ##### Against
Ha:The attributes IQ level and music are independent

```python
# import the required library
from scipy.stats import chi2_contingency

#write the data
frequencies=np.array([[45,32,13],[58,62,30],[87,126,87]])

#perform the test
statistics,p,dof,expected= chi2_contingency(frequencies)

#print the test statistics and the p-value
statistics,p
```

P-value(0.001) is less than 0.05 ,Hence we reject Ho

### Test for goodness of fit

#### Question:

A marketing manager at Apple inc. claims that silver colored ipad's have 43% demand,Space grey have 48% demand and gold colored have only 9% demand.

For a sample of 560 buyers of ipads,the color of device was recorded.

```python
Image(filename='D:/Imarticus/Probablity & Statics/analysis of variance/Chi_square goodness_question.PNG',width=400, height=200)
```

Test the managers claim.

#### Solution:

The hypothesis is:
Ho:There is no difference between observed frequencies and expected frequencies from the specified distribution

##### Against

Ha:There is difference between observed frequencies and expected frequencies from the specified distribution

The Specified distributio is the marketing managers claim.


```python
# import the required library
from scipy.stats import chisquare

#observed frequencies are
observed_frequencies=[280,190,90]
 
#expected frequencies are
#multiply the chance of each color by sample size in order to obtain the frequencies
expected_frequencies=[560*0.43,560*0.48,560*0.09]

#perform the test
chisquare(observed_frequencies,expected_frequencies)
```

p-value(0.000)is less than 0.005, Hence we reject Ho
